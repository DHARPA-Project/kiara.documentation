```python
kiara.list_operation_ids('download')
```

```python
#topic modelling this time so want a bigger corpus (also opportunity to introduce file_bundle rather than single file)

kiara.retrieve_operation_info('download.file_bundle')
```

```python
inputs = {
    "url": "https://github.com/DHARPA-Project/kiara.examples/archive/refs/heads/main.zip",
    "sub_path": "kiara.examples-main/examples/data/text_corpus/data"
 }

outputs = kiara.run_job('download.file_bundle', inputs=inputs)
outputs
```

```python
kiara.retrieve_operation_info('create.table.from.file_bundle')
```

```python
downloaded_files = outputs['file_bundle']

inputs = {
    'file_bundle' : downloaded_files
}

outputs = kiara.run_job('create.table.from.file_bundle', inputs=inputs)
outputs
```

```python
inputs = {
    'table' : outputs['table'],
    'column_name' : 'content'
}

outputs_contents = kiara.run_job('table.pick.column', inputs=inputs)
outputs_contents
```

```python
inputs = {
    'table' : outputs['table'],
    'column_name' : 'file_name'
}

outputs_file_name = kiara.run_job('table.pick.column', inputs=inputs)
outputs_file_name
```

```python
inputs = {
    'array' : outputs_file_name['array'],
    'min_index': 11,
    'max_index': 21
}

outputs_dates = kiara.run_job('parse.date_array', inputs=inputs)
outputs_dates
```

```python
#Now having downloaded our material for text analysis, we want to see what modules are included in the language processing plugin

infos = metadata = kiara.retrieve_operations_info()
operations = {}
for op_id, info in infos.item_infos.items():
    if info.context.labels.get("package", None) == "kiara_plugin.language_processing":
        operations[op_id] = info

print(operations.keys())
```
